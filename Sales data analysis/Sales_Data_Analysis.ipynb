```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
sales_df = pd.read_csv("sample_sales_data_100.csv")  # Update with the correct path if necessary
```

```python
# Calculate sales per transaction
sales_df['Sales'] = sales_df['Quantity'] * sales_df['Price']

# Calculate total sales
total_sales = sales_df['Sales'].sum()
```

```python
# Most Sold Product
product_sales = sales_df.groupby('Product')['Quantity'].sum()
most_sold_product = product_sales.idxmax()
most_sold_quantity = product_sales.max()
```

```python
# Sales by Region
sales_by_region = sales_df.groupby('Region')['Sales'].sum()
```

```python
print("Total Sales:", total_sales)


print("Most Sold Product:", most_sold_product)
print("Quantity Sold:", most_sold_quantity)


print("Sales by Region:")
print(sales_by_region)

# Visualization: Sales by Region (Pie Chart)
plt.figure(figsize=(10, 5))
sales_by_region.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title("Sales Distribution by Region")
plt.ylabel("")  # Remove y-axis label for a cleaner look

plt.show()

# Visualization: Most Sold Product (Bar Chart)
plt.figure(figsize=(10, 5))
product_sales.plot(kind='bar', color='skyblue')
plt.title("Quantity Sold per Product")
plt.xlabel("Product")
plt.ylabel("Quantity Sold")
```
